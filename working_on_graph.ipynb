```python
def check_nx_graphs_equal(g1, g2):
    sorted_g1_edges = sorted(g1.edges(data=True))
    sorted_g2_edges = sorted(g2.edges(data=True))
    equality = True
    for i in range(len(g1.edges())):
        equality = equality and  sorted_g1_edges[i][0] == sorted_g2_edges[i][0] 
        equality = equality and sorted_g1_edges[i][1] == sorted_g2_edges[i][1]
        for key, value in sorted_g1_edges[i][2].items():
            equality = equality and sorted_g1_edges[i][2][key] == sorted_g2_edges[i][2][key] 
    for n1, n2 in zip(sorted(g1.nodes(data=True)), sorted(g2.nodes(data=True))):
        equality = equality and (n1[0] == n2[0])
    return equality
```

```python

def print_2d_array(arr, width=5):
    for row in arr:
        print("\t[", end='')
        for element in row:
            print(f"{element:>{width}}", end=' ')
        print("]")


# Example usage
array_2d = [
    [1, 23, 456],
    [7890, 12, 345],
    [678, 901, 2345]
]
print_2d_array(array_2d)
```
